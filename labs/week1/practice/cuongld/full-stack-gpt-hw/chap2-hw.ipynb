```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.callbacks import StreamingStdOutCallbackHandler
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import BaseOutputParser

chat = ChatOpenAI(
    base_url=os.getenv("OPENAI_BASE_URL"),
    api_key=os.getenv("OPENAI_API_KEY"),
    model="gpt-5.1",
    temperature=0.1,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)
```

```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate, ChatPromptTemplate

examples = [
    {
        "input": "We sincerely appreciate your cooperation and look forward to continuing our partnership.",
        "answer": "귀하의 협조에 진심으로 감사드리며, 앞으로도 지속적인 협력을 기대합니다.",
    },
        {
        "input": "I'm really tired today, but I still need to finish my work.",
        "answer": "오늘 진짜 피곤한데, 그래도 일은 끝내야 해.",
    },
        {
        "input": "The sunset painted the sky with shades of gold and crimson.",
        "answer": "석양이 하늘을 황금빛과 진홍빛으로 물들였다.",
    },
]

example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "{input}"),
        ("ai", "{answer}"),
    ]
)

example_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
)

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a professional translator. Translate the given text into natural Korean while preserving the original tone and style as shown in the example."),
        example_prompt,
        ("human", "{input}"),
    ]
)
     
chain = final_prompt | chat
chain.invoke({"input": "Nice to see you!"})
```

```python
from langchain_core.globals import set_llm_cache
from langchain_community.cache import SQLiteCache
from langchain_core.callbacks import get_usage_metadata_callback


set_llm_cache(SQLiteCache("cache.db"))

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant that translates English to Korean."),
        ("human", "{input}"),
    ]
)

chain = prompt | chat

message = {"input": "what is your father name?"}

with get_usage_metadata_callback() as callback:
    chain.invoke(message)
    print("usage_metadata:", callback.usage_metadata)

with get_usage_metadata_callback() as callback:
    chain.invoke(message)
    print("usage_metadata:", callback.usage_metadata)   
```
