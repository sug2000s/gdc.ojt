```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
import os

load_dotenv()

llm = ChatOpenAI(
    base_url=os.getenv("OPENAI_BASE_URL"),
    api_key=os.getenv("OPENAI_API_KEY"),
    model=os.getenv("OPENAI_MODEL_NAME", "gpt-5.1"),
    temperature=0.1,
)
```

```python
from langchain_classic.memory import ConversationBufferMemory

conversation_turns = [
    ("Hi, I'm Nicolas and I live in South Korea.", "Nice to meet you, Nicolas from South Korea."),
    ("I work as a software engineer.", "Got it. You work as a software engineer."),
    ("My favorite food is kimchi.", "Understood. Your favorite food is kimchi."),
    ("I have a dog named Choco.", "Okay. Your dog is named Choco."),
    ("I usually wake up at 6 AM.", "Noted. You usually wake up at 6 AM."),
    ("On weekends I like hiking.", "Understood. You like hiking on weekends."),
    ("I'm learning Spanish these days.", "Got it. You are learning Spanish these days."),
    ("My favorite soccer team is Tottenham.", "Okay. Your favorite soccer team is Tottenham."),
    ("I want to visit Japan next year.", "Noted. You want to visit Japan next year."),
    ("I prefer coffee over tea.", "Got it. You prefer coffee over tea."),
]

buffer_memory = ConversationBufferMemory(return_messages=False)

for human_text, ai_text in conversation_turns:
    buffer_memory.save_context({"input": human_text}, {"output": ai_text})

loaded = buffer_memory.load_memory_variables({"input": "What do you remember about me?"})
history = loaded.get("history", "")

print("=== ConversationBufferMemory ===")
print("Memory after turn 10:")
print(history)
print()
print("Remembers turn 1:", "Nicolas" in history and "South Korea" in history)
print("Memory token count:", llm.get_num_tokens(history))
```

```python
from langchain_classic.memory import ConversationSummaryBufferMemory

conversation_turns = [
    ("Hi, I'm Nicolas and I live in South Korea.", "Nice to meet you, Nicolas from South Korea."),
    ("I work as a software engineer.", "Got it. You work as a software engineer."),
    ("My favorite food is kimchi.", "Understood. Your favorite food is kimchi."),
    ("I have a dog named Choco.", "Okay. Your dog is named Choco."),
    ("I usually wake up at 6 AM.", "Noted. You usually wake up at 6 AM."),
    ("On weekends I like hiking.", "Understood. You like hiking on weekends."),
    ("I'm learning Spanish these days.", "Got it. You are learning Spanish these days."),
    ("My favorite soccer team is Tottenham.", "Okay. Your favorite soccer team is Tottenham."),
    ("I want to visit Japan next year.", "Noted. You want to visit Japan next year."),
    ("I prefer coffee over tea.", "Got it. You prefer coffee over tea."),
]

summary_memory = ConversationSummaryBufferMemory(
    llm=llm,
    max_token_limit=100,
    return_messages=False,
)

for human_text, ai_text in conversation_turns:
    summary_memory.save_context({"input": human_text}, {"output": ai_text})

loaded = summary_memory.load_memory_variables({"input": "What do you remember about me?"})
history = loaded.get("history", "")

print("=== ConversationSummaryBufferMemory(max_token_limit=100) ===")
print("Memory after turn 10:")
print(history)
print()
print("Remembers turn 1:", "Nicolas" in history and "South Korea" in history)
print("Memory token count:", llm.get_num_tokens(history))
```
