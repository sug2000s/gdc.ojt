```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

import os

load_dotenv()

chat = ChatOpenAI(
    base_url = os.getenv("OPENAI_BASE_URL"),
    api_key = os.getenv("OPENAI_API_KEY"), 
    model = os.getenv("OPENAI_MODEL_NAME", "gpt-5.1"),
    temperature = 0.1
)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

class OutputParser(StrOutputParser):
    def parse(self, text: str) -> str:
        return text.strip()

template = ChatPromptTemplate.from_messages([
    ("system", "You are a professional translator. You will be given a text in {source_language}, and you need to translate it to {target_language}."), 
    ("human", "{text}")
])

chain = template | chat | OutputParser()

chain.invoke({
    "source_language": "English",
    "target_language": "Korean",
    "text": "Hello, how are you?"
})   
```
